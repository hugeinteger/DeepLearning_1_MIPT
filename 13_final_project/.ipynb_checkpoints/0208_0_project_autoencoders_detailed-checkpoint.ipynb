Осталось привести картинки к тензорам из PyTorch, чтобы можно было потом скармливать их автоэнкодеру: 

```python
class MyDataset(Dataset):
    def __init__(self, ds):
        super().__init__()
        # dataset
        self.ds = ds
        self.len_ = len(self.ds)
                      
    def __len__(self):
        return self.len_
  
    def __getitem__(self, index):
        # transform to torch tensor (CxHxW) and make elements to be from [0, 1]
        transform = transforms.Compose([
            transforms.ToTensor(),
        ])
        x = self.ds[index]
        x = np.array(x / 255, dtype='float32')
        x = transform(x)
        return x
```

## Autoencoder

Why to use all this complicated formulaes and regularizations, what is the need for variational inference? To analyze the difference, let's first train just an autoencoder on the data:
<img src="https://i.imgur.com/nVJAtMT.png" alt="Autoencoder">

```python
inp_size=X_train.shape[1]
# размерность промежуточных слоев 
hid_size=250
# размерность среднего слоя - latent dimension
dimZ = 256
```

Реализуем autoencoder. Архитектуру (conv, fully-connected, ReLu, etc) можете выбирать сами. Экспериментируйте!

```python
class Encoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.encoder_conv = nn.Sequential(
                     nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=16),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3),
                     nn.BatchNorm2d(num_features=32),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=64),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3),
                     nn.BatchNorm2d(num_features=128),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=256),
                     nn.ReLU(),
      )
      self.encoder_lin = nn.Sequential(
                    nn.Linear(256*3*3, dimZ),
                    nn.BatchNorm1d(num_features=dimZ),
                    nn.ReLU()
      )

    def forward(self, x):
      x = self.encoder_conv(x)
      x = x.view(-1, 256*3*3)
      latent_code = self.encoder_lin(x)

      return latent_code
```

```python
class Decoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.decoder_lin = nn.Sequential(
                    nn.Linear(dimZ, 256*3*3),
                    nn.BatchNorm1d(num_features=256*3*3),
                    nn.ReLU()
      )
      self.decoder_deconv = nn.Sequential(
                     nn.ConvTranspose2d(in_channels=256, out_channels=128, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=128),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=128, out_channels=64, kernel_size=3),
                     nn.BatchNorm2d(num_features=64),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=64, out_channels=32, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=32),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=32, out_channels=16, kernel_size=3),
                     nn.BatchNorm2d(num_features=16),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=16, out_channels=8, kernel_size=3, stride=2),
                     nn.BatchNorm2d(num_features=8),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=8, out_channels=3, kernel_size=3),
                     nn.Sigmoid()
      )

    def forward(self, x):
      x = self.decoder_lin(x)
      x = x.view(-1, 256, 3, 3)
      reconstruction = self.decoder_deconv(x)
      return reconstruction
```

```python
class Autoencoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.encoder = Encoder()
      self.decoder = Decoder()
    def forward(self, x):
      latent_code = self.encoder(x)
      reconstruction = self.decoder(latent_code)
      return latent_code, reconstruction
```

Осталось написать код обучения автоэнкодера. При этом было бы неплохо в процессе иногда смотреть, как автоэнкодер реконструирует изображения на данном этапе обучения. Наример, после каждой эпохи (прогона train выборки через автоэекодер) можно смотреть, какие реконструкции получились для каких-то изображений val выборки.

Подсказка: если x_val -- каринка, а reconstruction -- ее реконструкция автоэнкодером, то красиво вывести эту каритинку и ее реконструкцию можно с помощью функции plot_gallery вот так:

*plot_gallery([x_val, reconstruction], image_h, image_w, n_row=1, n_col=2)*

А, ну еще было бы неплохо выводить графики train и val лоссов в процессе тренировки =)

```python
def fit_epoch(model, train_loader, criterion, optimizer):
    running_loss = 0.0
    processed_data = 0
  
    for inputs in train_loader:
        inputs = inputs.to(DEVICE)
        optimizer.zero_grad()

        latents, outputs = model(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()*inputs.size(0)
        processed_data += inputs.size(0)
              
    train_loss = running_loss / processed_data
    return train_loss
```

```python
def eval_epoch(model, val_loader, criterion):
    model.eval()
    running_loss = 0.0
    processed_size = 0

    for batch_idx, inputs in enumerate(val_loader):
        inputs = inputs.to(DEVICE)

        with torch.set_grad_enabled(False):
            latents, outputs = model(inputs)
            loss = criterion(outputs, inputs)

        running_loss += loss.item() * inputs.size(0)
        processed_size += inputs.size(0)
        # plot some images
        if batch_idx % 1000 == 1:
          plot_gallery([inputs[0].transpose(0,2).transpose(0,1).cpu(), outputs[0].transpose(0,2).transpose(0,1).cpu()], IMAGE_H, IMAGE_W, 3, 1, 2)
    val_loss = running_loss / processed_size
    return val_loss
```

```python
def train(tr, val, model, epochs, batch_size):
    train_loader = DataLoader(tr, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val, batch_size=batch_size, shuffle=False)

    history = []
    log_template = "\nEpoch {ep:03d} train_loss: {t_loss:0.4f} \
    val_loss {v_loss:0.4f}"

    with tqdm(desc="epoch", total=epochs) as pbar_outer:
        criterion = nn.MSELoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

        for epoch in range(epochs):
            train_loss = fit_epoch(model, train_loader, criterion, optimizer)
            print("loss", train_loss)
            
            val_loss = eval_epoch(model, val_loader, criterion)
            history.append((train_loss, val_loss))
            
            pbar_outer.update(1)
            tqdm.write(log_template.format(ep=epoch+1, t_loss=train_loss,\
                                           v_loss=val_loss))
            
    return history
```

```python
autoencoder = Autoencoder().to(DEVICE)
history = train(train_dataset, val_dataset, model=autoencoder, epochs=100, batch_size=32)

loss, val_loss = zip(*history)

plt.figure(figsize=(15, 9))
plt.plot(loss, label="train_loss")
plt.plot(val_loss, label="val_loss")
plt.legend(loc='best')
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

Давайте посмотрим, как наш тренированный автоэекодер кодирует и восстанавливает картинки:

```python
def predict_one_sample(model, inputs, device=DEVICE):
    """Предсказание, для одной картинки"""
    with torch.no_grad():
        inputs = inputs.to(device)
        model.eval()
        latent, pred = model(inputs)
    return latent, pred
```

```python
for i in range(20):
  random_characters = int(np.random.uniform(0,1000))
  ex_img = val_dataset[random_characters]
  latent, reconstr = predict_one_sample(autoencoder, ex_img.unsqueeze(0))

  ex_img = ex_img.transpose(0,2).transpose(0,1)
  reconstr = reconstr.squeeze(0).transpose(0,2).transpose(0,1)

  plot_gallery([ex_img.cpu(), reconstr.cpu()], IMAGE_H, IMAGE_W, 1, 2)
```

## Sampling

Давайте теперь будем не просто брать картинку, прогонять ее через автоэекодер и получать реконструкцию, а попробуем создать что-то НОВОЕ

Давайте возьмем и подсунем декодеру какие-нибудь сгенерированные нами векторы (например, из нормального распределения) и посмотрим на результат реконструкции декодера:

#### If that doesn't work
Если вместо лиц у вас выводится непонятно что, попробуйте посмотреть, как выглядят латентные векторы картинок из датасета. Так как в обучении нейронных сетей есть определенная доля рандома, векторы латентного слоя могут быть распределены НЕ как np.random.randn(25, <latent_space_dim>). А чтобы у нас получались лица при запихивании вектора декодеру, вектор должен быть распределен так же, как лаьентные векторы реальных фоток. Так что ридется рандом подогнать.

```python
# let's get latent vectors of our validation set
def get_latents(my_dataset):
  latents = np.empty((0, dimZ))
  val_loader = DataLoader(my_dataset, batch_size=1, shuffle=False)
  for inputs in val_loader:
    latent, reconstr = predict_one_sample(autoencoder, inputs)
    latents = np.append(latents, [latent.flatten().cpu().numpy()], axis=0)
  return latents
```

```python
# create random 25 vectors of dimZ size and make them to have mean and std (smaller) of our train set latetnts
latents = get_latents(train_dataset)
z = np.random.normal(loc=latents.mean(axis=0), scale=0.3*latents.std(axis=0), size=(25, dimZ))

with torch.no_grad():
    z = torch.from_numpy(z).float().to(DEVICE)
    autoencoder.decoder.eval()
    output = autoencoder.decoder(z)
    output = output.transpose(1,3).transpose(1,2)
plot_gallery(output.data.cpu().numpy(), IMAGE_H, IMAGE_W, 3, n_row=5, n_col=5)
```

Давайте теперь допустим, что мы обучили AE на датасете MNIST и теперь хотим генерировать новые картинки с числами с помощью декодера (как выше мы генерили рандомные лица). 
И вот мне понадобилось сгенерировать цифру 8. И я подставляю разные варианты шума, и все никак не генерится восьмерка -- у меня получаются то пятерки, то тройки, то четверки. Гадость(

  Типа я такая говорю "выдай мне случайную восьмерку" и оно генерит случайную восьмерку!

Conditional AE -- так называется вид автоэнкодера, который предоставляет такую возможность. Ну, название "conditional" уже говорит само за себя.

И в этой части проекта мы научимся такие обучать.

## Архитектура

На картинке ниже представлена архитектура простого Conditional AE.

По сути, единственное отличие от обычного -- это то, что мы вместе с картинкой в первом слое энкодера и декодера передаем еще информацию о классе картинки. 

То есть, в первый (входной) слой энкодера есть конкатенация картинки и информации о классе (например, вектора из девяти нулей и одной единицы). Первый слой декодера есть конкатенация латентного вектора и информации о классе.


![alt text](https://i.ibb.co/2tsWknB/Screen-Shot-2020-01-15-at-9-02-15-PM.png)



Таким образом, при генерации новой рандомной картинки мы должны будем передать декодеру сконкатенированные латентный вектор и класс картинки.

### P.S.
Можно ередавать класс картинки не только в первый слой, но и в каждый слой сети. То есть на каждом слое конкатенировать выход из предыдущего слоя и информацию о классе.

```python
# initialize globals

# we're going to run 10 epochs
EPOCHS = 10
# number of classes (=our one-hot encoding dimension)
N_CLASSES = 10
# hidden layers dimensions
DIM_HIDDEN = 256
# latent dimension
DIM_Z = 75
# input item size
INPUT_SIZE = 28 * 28
# работаем на видеокарте
DEVICE = torch.device("cuda")
```

```python
# one-hot encoding, numbers - tensor of BATCH_SIZE 
def num2onehot(numbers, n=N_CLASSES):
    onehot = torch.zeros(numbers.size(0), n)
    onehot.scatter_(1, numbers.data, 1)

    return onehot
```

```python
# conditional encoder
class CondEncoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.lin1 = nn.Sequential(
                     nn.Linear(INPUT_SIZE + N_CLASSES, INPUT_SIZE),
                     nn.ReLU()
      )
      self.encoder_conv = nn.Sequential(
                     nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3), # BSx8x26x26
                     nn.BatchNorm2d(num_features=8),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, stride=2), # BSx16x12x12
                     nn.BatchNorm2d(num_features=16),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3), # BSx32x10x10
                     nn.BatchNorm2d(num_features=32),
                     nn.ReLU(),
                     nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=2), # BSx64x4x4
                     nn.BatchNorm2d(num_features=64),
                     nn.ReLU(),
      )
      self.lin2 = nn.Sequential(
                    nn.Linear(64*4*4+N_CLASSES, 512),
                    nn.BatchNorm1d(num_features=512),
                    nn.ReLU()
                    # nn.Linear(512, 256),
                    # nn.BatchNorm1d(num_features=256),
                    # nn.ReLU(),
                    # nn.Linear(256, 128),
                    # nn.BatchNorm1d(num_features=128),
                    # nn.ReLU()
      )
      self.lin3 = nn.Sequential(
                    nn.Linear(512+N_CLASSES, DIM_Z),
                    nn.BatchNorm1d(num_features=DIM_Z),
                    nn.ReLU()
      )

    def forward(self, x, y):
      # x: BATCH_SIZExINPUT_SIZE, y: BATCH_SIZExN_CLASSES
      # add one-hot encoding of the class
      x = torch.cat((x, y), dim=1) # BATCH_SIZEx(INPUT_SIZE+N_CLASSES)

      x = self.lin1(x) # BATCH_SIZExINPUT_SIZE
      # flatten
      x = x.view(-1, 1, 28, 28) # BATCH_SIZEx1x28x28

      x = self.encoder_conv(x) # BSx64x4x4
      # flatten
      x = x.view(-1, 64*4*4)
      # add one-hot encoding of the class
      x = torch.cat((x, y), dim=1) # BATCH_SIZEx(64*4*4+N_CLASSES)
      x = self.lin2(x) # BATCH_SIZEx128

      # add one-hot encoding of the class
      x = torch.cat((x, y), dim=1) # BATCH_SIZEx(128+N_CLASSES)
      latent_code = self.lin3(x) # BATCH_SIZExDIM_Z

      return latent_code
```

```python
class CondDecoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.lin3 = nn.Sequential(
                    nn.Linear(DIM_Z + N_CLASSES, 512),
                    nn.BatchNorm1d(num_features=512),
                    nn.ReLU()
      )
      self.lin2 = nn.Sequential(
                    # nn.Linear(128, 256),
                    # nn.BatchNorm1d(num_features=256),
                    # nn.ReLU(),
                    # nn.Linear(256, 512),
                    # nn.BatchNorm1d(num_features=512),
                    # nn.ReLU(),
                    nn.Linear(512, 64*4*4),
                    nn.BatchNorm1d(num_features=64*4*4),
                    nn.ReLU(),
      )
      self.decoder_deconv = nn.Sequential(
                     nn.ConvTranspose2d(in_channels=64, out_channels=32, kernel_size=4, stride=2), # BSx32x10x10
                     nn.BatchNorm2d(num_features=32),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=32, out_channels=16, kernel_size=3), # BSx16x12x12
                     nn.BatchNorm2d(num_features=16),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=16, out_channels=8, kernel_size=3, stride=2), # BSx8x25x25
                     nn.BatchNorm2d(num_features=8),
                     nn.ReLU(),
                     nn.ConvTranspose2d(in_channels=8, out_channels=1, kernel_size=4), #BSx1X28x28
                     nn.Sigmoid()
      )

    def forward(self, x, y):
      # x: BATCH_SIZExDIM_Z, y: BATCH_SIZExN_CLASSES
      # add one-hot encoding of the class
      x = torch.cat((x, y), dim=1) # BATCH_SIZEx(DIM_Z+N_CLASSES)
      x = self.lin3(x) # BATCH_SIZEx128

      x = self.lin2(x) # BATCH_SIZEx(64*4*4)

      # reshape
      x = x.view(-1, 64, 4, 4) # BATCH_SIZEx64x4x4
      # get the image
      reconstruction = self.decoder_deconv(x) #BATCH_SIZEx1x28x28

      return reconstruction
```

```python
# conditional autoencoder
class CondAutoencoder(nn.Module):
    def __init__(self):
      super().__init__()
      self.encoder = CondEncoder()
      self.decoder = CondDecoder()
    def forward(self, x, y):
      # encode
      latent_code = self.encoder(x, y)
      # decode
      reconstruction = self.decoder(latent_code, y)
      return latent_code, reconstruction
```

```python
def fit_epoch(model, train_loader, criterion, optimizer):
    # print("in fit_epoch")
    running_loss = 0.0
    processed_data = 0
  
    for (inputs, labels) in train_loader:
        # reshape the data into [BATCH_SIZE, INPUT_SIZE]
        # print("in fit epock in", inputs.shape)
        inputs_rs = inputs.view(-1, INPUT_SIZE)
        inputs_rs = inputs_rs.to(DEVICE)
        inputs = inputs.to(DEVICE)

        # convert labels into one-hot encoding
        labels = num2onehot(labels.view(-1, 1))
        labels = labels.to(DEVICE)

        optimizer.zero_grad()

        latents, outputs = model(inputs_rs, labels)
        # print("in fit epock ", inputs.shape, outputs.shape)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()*inputs.size(0)
        processed_data += inputs.size(0)
              
    train_loss = running_loss / processed_data
    return train_loss
```

```python
def eval_epoch(model, val_loader, criterion):
    model.eval()
    running_loss = 0.0
    processed_size = 0

    for batch_idx, (inputs, labels) in enumerate(val_loader):
        # reshape the data into [BATCH_SIZE, INPUT_SIZE]
        inputs_rs = inputs.view(-1, INPUT_SIZE)
        inputs_rs = inputs_rs.to(DEVICE)
        inputs = inputs.to(DEVICE)

        # convert labels into one-hot encoding
        labels = num2onehot(labels.view(-1, 1))
        labels = labels.to(DEVICE)

        with torch.set_grad_enabled(False):
            latents, outputs = model(inputs_rs, labels)
            loss = criterion(outputs, inputs)

        running_loss += loss.item() * inputs.size(0)
        processed_size += inputs.size(0)
        # plot some images
        if batch_idx % 1000 == 1:
          plot_gallery([inputs[0], outputs[0]], 28, 28, 1, 2)
          val_loss = running_loss / processed_size
    return val_loss
```

```python
def train(train_loader, test_loader, model, epochs):

    history = []
    log_template = "\nEpoch {ep:03d} train_loss: {t_loss:0.4f} \
    val_loss {v_loss:0.4f}"

    with tqdm(desc="epoch", total=epochs) as pbar_outer:
        criterion = nn.MSELoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
        for epoch in range(epochs):
            train_loss = fit_epoch(model, train_loader, criterion, optimizer)
            print("loss", train_loss)
            
            val_loss = eval_epoch(model, test_loader, criterion)
            history.append((train_loss, val_loss))
            
            pbar_outer.update(1)
            tqdm.write(log_template.format(ep=epoch+1, t_loss=train_loss,\
                                           v_loss=val_loss))
            
    return history
```

```python
condAutoencoder = CondAutoencoder().to(DEVICE)
# print(condAutoencoder)

history = train(train_loader, test_loader, model=condAutoencoder, epochs=60)

loss, val_loss = zip(*history)

plt.figure(figsize=(15, 9))
plt.plot(loss, label="train_loss")
plt.plot(val_loss, label="val_loss")
plt.legend(loc='best')
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
def KL_divergence(mu, logsigma):
    """
    часть функции потерь, которая отвечает за "близость" латентных представлений разных людей
    """
    return -(1/2) * (1 + 2*logsigma - mu**2 - torch.exp(logsigma)**2).sum(dim=-1)

def log_likelihood(x, reconstruction):
    """
    часть функции потерь, которая отвечает за качество реконструкции (как mse в обычном autoencoder)
    """
    loss = nn.BCELoss()
    return loss(reconstruction, x)

def loss_vae(x, mu, logsigma, reconstruction):
    return -(-KL_divergence(mu, logsigma) + log_likelihood(x, reconstruction)).mean()
```
